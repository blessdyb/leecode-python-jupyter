```python
class Solution:
    def zigzagConversion(self, s: str, numRows: int) -> str:
        # Edge case
        if numRows < 2:
            return s
        # Use list to create unknown length two dimension array
        matrix = [[] for _ in range(numRows)]
        j = 0
        direction = 1
        for i in range(len(s)):
            if j == -1:
                j = 1
                direction = 1
            elif j == numRows:
                j = numRows - 2
                direction = -1
            matrix[j].append(s[i])
            if direction == 1:
                j += 1
            else:
                j -= 1
        return ''.join([''.join(item) for item in matrix])

    def romanToInt(self, s: str) -> int:
        numLength = len(s)
        hashes = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        i = 0
        result = 0
        while i < numLength:
            if i + 1 >= numLength:
                result += hashes[s[i]]
                i += 1
            elif s[i] == 'I' and s[i + 1] == 'V':
                result += 4
                i += 2
            elif s[i] == 'I' and s[i + 1] == 'X':
                result += 9
                i += 2
            elif s[i] == 'X' and s[i + 1] == 'L':
                result += 40
                i += 2
            elif s[i] == 'X' and s[i + 1] == 'C':
                result += 90
                i += 2
            elif s[i] == 'C' and s[i + 1] == 'D':
                result += 400
                i += 2
            elif s[i] == 'C' and s[i + 1] == 'M':
                result += 900
                i += 2
            else:
                result += hashes[s[i]]
                i += 1
        return result
```
